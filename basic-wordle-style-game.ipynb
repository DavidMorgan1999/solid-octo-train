```python
df = pd.read_csv("../input/wordle-5-letter-words/5_letters.csv")
df.head()
```

```python
word_choice = df.sample(n=1)
answer_list = word_choice.values.flatten().tolist()
answer_list
```

```python
def Guess(word):
    while len(word) != 5:
        word = input("Input a 5 letter word:")
    word.lower()
    output = [char for char in word]
    return output
```

```python
#guessed_word = Guess(input("Guess: "))
guessed_word = Guess("Beans")
score_guess(guessed_word)
```

```python
def score_guess(word):
    i = 0
    word_score = []
    for x in word:
        letter = [x,0]
        if x in answer_list:
            letter = [x,1]
        if x == answer_list[i]:
            letter = [x,2]
        i = i+1
        word_score.append(letter)
    return word_score
```
